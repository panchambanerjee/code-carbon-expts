```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df_bert_cased_a100 = pd.read_csv('emissions_bert-base-cased-epochs-1-2-5-10_A100.csv')
df_bert_cased_a100
```

```python
df_bert_uncased_t4 = pd.read_csv('emissions_bert-base-uncased-epochs-1-2-5-10_T4.csv')
df_bert_uncased_t4
```

```python
df_bert_multi_cased_t4 = pd.read_csv('emissions_bert-base-multilingual-cased-epochs-1-2-5-10_T4.csv')
df_bert_multi_cased_t4
```

```python
df_bert_multi_cased_v100 = pd.read_csv('emissions_bert-base-multilingual-cased-epochs-1-2-5-10_V100.csv')
df_bert_base_uncased_v100 = pd.read_csv('emissions_bert-base-uncased-epochs-1-2-5-10_V100.csv')
df_bert_base_cased_v100 = pd.read_csv('emissions_bert-base-cased-epochs-1-2-5-10_V100.csv')
df_bert_base_cased_t4 = pd.read_csv('emissions_bert-base-cased-epochs-1-2-5-10_T4.csv')
```

```python
df_bert_cased_a100['epochs'] = [1,2,5,10]
df_bert_uncased_t4['epochs'] = [1,2,5,10]
df_bert_multi_cased_t4['epochs'] = [1,2,5,10]
df_bert_multi_cased_v100['epochs'] = [1,2,5,10]
df_bert_base_cased_v100['epochs'] = [1,2,5,10]
df_bert_base_uncased_v100['epochs'] = [1,2,5,10]
df_bert_base_cased_t4['epochs'] = [1,2,5,10]
```

```python
df_bert_cased_a100
```

```python
sns.lineplot(x=df_bert_cased_a100['epochs'], y=df_bert_cased_a100['emissions'])
```

```python
# Set the style and font sizes
sns.set_style('ticks')
plt.rcParams.update({'font.size': 8})
 
# Create a Seaborn line plot with different markers for each product
sns.lineplot(x='epochs', y='emissions', data=df_bert_multi_cased_t4, marker='o', color='navy', label='Base Multilingual Cased', alpha=.6)
sns.lineplot(x='epochs', y='emissions', data=df_bert_uncased_t4, marker='s', color='teal', label='Base Cased', alpha=.6)
# sns.lineplot(x='Months', y='Product C', data=df, marker='D', color='salmon', label='Product C')
 
# Set plot title and axes labels
plt.title('Emissions across different BERT Models (T4 trained')
plt.xlabel('Epochs')
plt.ylabel('Emissions')
 
# Add a legend
plt.legend(loc='lower right')
 
# Add a grid
plt.grid(True)
 
# Remove the top and right spines
sns.despine()
 
# Show the plot
plt.show()
```

```python
# Set the style and font sizes
plt.figure(figsize=(10,8))

sns.set_style('ticks')
plt.rcParams.update({'font.size': 8})
 
# Create a Seaborn line plot with different markers for each product
sns.lineplot(x='epochs', y='emissions', data=df_bert_multi_cased_v100, marker='o', color='navy', \
             label='Base Multilingual Cased (V100)', alpha=.6)
sns.lineplot(x='epochs', y='emissions', data=df_bert_base_cased_v100, marker='s', color='greenyellow', \
             label='Base Cased (V100)', alpha=.6)
sns.lineplot(x='epochs', y='emissions', data=df_bert_base_uncased_v100, marker='D', color='salmon', \
             label='Base Uncased (V100)', alpha=.6)
# sns.lineplot(x='Months', y='Product C', data=df, marker='D', color='salmon', label='Product C')

sns.lineplot(x='epochs', y='emissions', data=df_bert_multi_cased_t4, marker='>', color='blue', \
             label='Base Multilingual Cased (T4)', alpha=.6)
sns.lineplot(x='epochs', y='emissions', data=df_bert_base_cased_t4, marker='^', color='green', \
             label='Base Cased (T4)', alpha=.6)
sns.lineplot(x='epochs', y='emissions', data=df_bert_uncased_t4, marker='<', color='red', \
             label='Base Uncased (T4)', alpha=.6)
 
# Set plot title and axes labels
plt.title('Emissions across different BERT Models (T4 and V100)', fontsize=16)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Emissions (CO2eq)', fontsize=12)
 
# Add a legend
plt.legend(loc='best')
 
# Add a grid
plt.grid(True)
 
# Remove the top and right spines
sns.despine()
 
# Show the plot
plt.show()
```

```python
# Set the style and font sizes
plt.figure(figsize=(10,8))

sns.set_style('ticks')
plt.rcParams.update({'font.size': 8})
 
# Create a Seaborn line plot with different markers for each product
sns.lineplot(x='epochs', y='duration', data=df_bert_multi_cased_v100, marker='o', color='navy', \
             label='Base Multilingual Cased (V100)', alpha=.6)
sns.lineplot(x='epochs', y='duration', data=df_bert_base_cased_v100, marker='s', color='greenyellow', \
             label='Base Cased (V100)', alpha=.6)
sns.lineplot(x='epochs', y='duration', data=df_bert_base_uncased_v100, marker='D', color='salmon', \
             label='Base Uncased (V100)', alpha=.6)
# sns.lineplot(x='Months', y='Product C', data=df, marker='D', color='salmon', label='Product C')

sns.lineplot(x='epochs', y='duration', data=df_bert_multi_cased_t4, marker='>', color='blue', \
             label='Base Multilingual Cased (T4)', alpha=.6)
sns.lineplot(x='epochs', y='duration', data=df_bert_base_cased_t4, marker='^', color='green', \
             label='Base Cased (T4)', alpha=.6)
sns.lineplot(x='epochs', y='duration', data=df_bert_uncased_t4, marker='<', color='red', \
             label='Base Uncased (T4)', alpha=.6)
 
# Set plot title and axes labels
plt.title('Durations across different BERT Models (T4 and V100)', fontsize=16)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Durations (s)', fontsize=12)
 
# Add a legend
plt.legend(loc='best')
 
# Add a grid
plt.grid(True)
 
# Remove the top and right spines
sns.despine()
 
# Show the plot
plt.show()
```
